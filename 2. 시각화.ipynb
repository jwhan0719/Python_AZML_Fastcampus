# 2. Subplot / Subplots 를 이용한 여러개의 차트 시각화

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

## 2.1 subplot을 이용
- subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
```

- tight_layout 을 이용하면 레이아웃이 자동으로 설정됨

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
plt.tight_layout()
```

## 2.2 subplots 를 이용

## 2.3 y축이 2개인 그래프

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color = 'red')
chart2 = axe2.plot(x2, y2, color = 'blue')

axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)

# 범례 표시
chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```

# 3. Barplot 활용한 막대 그래프 시각화

```python
import matplotlib.pyplot as plt
plt.rc("font", family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

import seaborn as sns
import pandas as pd
```

```python
movie_title = ['크루엘라', '극장판 귀멸의 칼날: 무한열차편', '학교 가는 길']
audience = [664308, 2099131, 20067]

data = {'영화제목' : movie_title, '누적관객' : audience}
df = pd.DataFrame(data)

sns.barplot(data = df, x = '영화제목', y = '누적관객')
```

## 3.1 차트 크기 변경

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 10,5
sns.barplot(data = df, x = '영화제목', y = '누적관객')
```

```python
sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객').영화제목)
```

## 3.2 차트 정렬

```python
# 누적관객수 오름차순 정렬
sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = True).영화제목)
```

```python
# 누적관객수 내림차순 정렬
sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = False).영화제목)
```

## 3.3 포맷 변경

```python
chart = sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = False).영화제목)

ylabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
```

## 3.4 가로 차트

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목)

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

## 3.5 차트 색상 변경

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목,
             color = 'blue')

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

## 3.6 제목 추가 및 폰트 사이즈 변경

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목,
             color = 'blue')

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)

plt.xlabel('누적관객', fontsize = 15)
plt.ylabel('영화제목', fontsize = 15)
plt.title('영화 별 누적관객수', fontsize = 20)
```

# 5. Scatterplot을 활용한 산점도 시각화

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# sepal_length, sepal_width 별 산점도
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length')
```

## 5.1 분류 설정

```python
# huie를 이용한 종 분류
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

- petal_length 의 값을 3 미만, 5 미만, 5 이상으로 분류 표시

```python
df['petal_length'].unique()
```

```python
def function1(x):
    if x < 3:
        return 's'
    elif x < 5:
        return 'm'
    else:
        return 'l'
    
df['petal_length2'] = df['petal_length'].apply(function1)
df.head()
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length2')
```

- 점 크기로 분류

```python
def function2(x):
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7

df['petal_length3'] = df['petal_length'].apply(function2)
df.head()
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length3')
```

## 5.2 카테고리별 그래프
- relplot

```python
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```

## 5.3 회귀선
- lmplot

```python
sns.lmplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

# 7. Heatmap 을 이용한 히트맵 시각화
스크린샷 2021-06-13 오후 12.40.39.png

```python
df = sns.load_dataset('flights')
df.head()
```

- 년도별 월별 승객수 히트맵 시각화

```python
pivot_data = df.pivot("month", "year", "passengers")
pivot_data
```

```python
sns.heatmap(pivot_data)
```

## 7.1 구분선 설정
- linewidths

```python
sns.heatmap(pivot_data, linewidths=.5)
```

## 7.2 색상 설정
- cmap

```python
sns.heatmap(pivot_data, cmap="Blues")
```

## 7.3 수치 설정
- annot=True

```python
sns.heatmap(pivot_data, cmap="Blues", annot = True)
```

- 정수형태 출력 -> fmt='d'

```python
sns.heatmap(pivot_data, cmap="Blues", annot = True, fmt = 'd')
```

## 7.4 가로 설정

```python
fig, (ax, cbar_ax) = plt.subplots(2)

ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "horizontal"})
```

- 수직 방향

```python
ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "vertical"})

fig
```

## 7.5 비율 조정
- colorbar와 히트맵 간의 비율 조정

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"},
                 cmap="Blues", 
                 annot = True, fmt = 'd')
```
