```python
# 두번째 행부터 네번째 행 까지 추출
df.loc[2:4]
```

```python
# 인덱스가 문자로 이뤄진 데이터프레임 생성
index = ['a', 'b', 'd', 'c', 'e', 'f', 'g', 'g', 'h', 'i']
df = pd.DataFrame({'a' : [i for i in range(1, 11)], 'b' : [i for i in range(11, 21)], 'c' : [i for i in range(21, 31)]}, index = index)

df
```

```python
# 열이 a,c이면서 인덱스가 g,i인 데이터 조회
df.loc[['g', 'i'], ['a', 'c']]
```

```python
df = pd.DataFrame({'a' : [i for i in range(1, 11)], 'b' : [i for i in range(11, 21)], 'c' : [i for i in range(21, 31)]})
df
```

```python
# 5번째 행까지 데이터 첫번째, 세번째 열 조회
df.iloc[:5, [0, 2]]
```

```python
# a가 3 이상인 데이터 중 a,c열만 조회
df[df['a'] >= 3][['a', 'c']]
```

```python
# a가 3 이상 이면서 b가 16 미만 데이터 조회
df[(df['a'] >= 3) & (df['b'] < 16)]
```

```python
# a가 3 이하 이거나 7 이상인 데이터 조회
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

```python
# a가 3 이상이고, b가 17미만이거나 c가 30인 데이터 조회
df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
```

# 7. 타입변환

```python
df = pd.DataFrame({'판매일' : ['5/11/21', '5/12/21', '5/13/21', '5/14/21', '5/15/21'],
                   '판매량' : ['10', '15', '20', '25', '30'], '방문자수' : ['10', '-', '17', '23', '25'], 
                   '기온' : ['24.1', '24.3', '24.8', '25', '25.4']})

df
```

```python
# 전체 컬럼 타입 확인
df.dtypes
```

```python
# 판매량 컬럼 정수타입으로 변환
print(df.dtypes, '\n')

df = df.astype({'판매량' : 'int'})
print(df.dtypes)
```

```python
# 방문자수를 숫자 타입으로 변경
# 방문자수 컬럼 데이터는 '-'문자를 포함하고 있어서 df.astype({'방문자수' : 'int'})으로 변경 불가능
print(df.dtypes, '\n')

# float형으로 변환
df['방문자수'] = pd.to_numeric(df['방문자수'], errors = 'coerce')
print(df.dtypes, '\n')

# 결측값 처리후 int형으로 변환
df.fillna(0, inplace = True)
df = df.astype({'방문자수' : 'int'})
print(df.dtypes)
```

```python
# 판매일 컬럼을 datetime 타입으로 변경
print(df.dtypes)
display(df)

df['판매일'] = pd.to_datetime(df['판매일'], format="%m/%d/%y")
print(df.dtypes)
display(df)
```

# 9. apply, map을 활용한 데이터 변환

### 9.1 apply 활용

```python
df = pd.DataFrame({'a' : [1, 2, 3, 4, 5]})
df
```

```python
# a 가 2 보다 작으면 '2 미만', 4 보다 작으면 '4 미만', 4 보다 크면 '4 이상' 이 저장된 b 칼럼을 추가
def case_function(x):
    if x < 2:
        return '2 미만'
    elif x < 4:
        return '4 미만'
    else:
        return '4 이상'
    
df['b'] = df['a'].apply(case_function)    
df
```

```python
# a 가 1 이면 'one', 2 이면 'two', 3 이면 'three', 4 이면 'four', 5 이면 'five' 를 출력하는 칼럼 추가
def function(x):
    if x == 1:
        return 'one'
    elif x == 2:
        return 'two'
    elif x == 3:
        return 'three'
    elif x == 4:
        return 'four'
    elif x == 5:
        return 'five'
    
df['c'] = df['a'].apply(function)
df
```

### 9.2 map 활용

```python
a = { 1 : 'one', 2 : 'two', 3 : 'three', 4 : 'four', 5 : 'five'}
df['d'] = df['a'].map(a)

df
```
